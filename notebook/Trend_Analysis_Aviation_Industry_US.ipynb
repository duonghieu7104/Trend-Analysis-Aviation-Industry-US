```python
import kagglehub
import pandas as pd

# Tải dataset về (sẽ trả về đường dẫn local)
path = kagglehub.dataset_download("bhavikjikadara/us-airline-flight-routes-and-fares-1993-2024")

print("Dataset downloaded to:", path)
```

```python
df = pd.read_csv(path + "/US Airline Flight Routes and Fares 1993-2024.csv")
print(df.shape)
df.head()
```

# EDA

```python
# Basic Data Overview
print("=== DATASET OVERVIEW ===")
print(f"Dataset shape: {df.shape}")
print(f"Number of rows: {df.shape[0]:,}")
print(f"Number of columns: {df.shape[1]}")
print("\n=== COLUMN INFORMATION ===")
print("Column names and data types:")
print(df.dtypes)
print(f"\nMemory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
# Missing Values Analysis
print("=== MISSING VALUES ANALYSIS ===")
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing_Count': missing_values.values,
    'Missing_Percentage': missing_percentage.values
}).sort_values('Missing_Percentage', ascending=False)

print("Missing values by column:")
print(missing_df[missing_df['Missing_Count'] > 0])

print(f"\nTotal missing values: {df.isnull().sum().sum():,}")
print(f"Percentage of missing values: {(df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100:.2f}%")
```

```python
# Basic Statistics for Numerical Columns
print("=== BASIC STATISTICS ===")
print("Numerical columns summary:")
print(df.describe())

print("\n=== CATEGORICAL COLUMNS ===")
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:")
for col in categorical_cols:
    unique_count = df[col].nunique()
    print(f"{col}: {unique_count} unique values")
    if unique_count <= 20:  # Show values if not too many
        print(f"  Values: {df[col].unique()}")
    print()
```

```python
# Data Quality Assessment - Duplicates
print("=== DUPLICATE ANALYSIS ===")
print(f"Total duplicate rows: {df.duplicated().sum()}")
print(f"Percentage of duplicates: {(df.duplicated().sum() / len(df)) * 100:.2f}%")

# Check for duplicates based on key columns
key_columns = ['Year', 'quarter', 'city1', 'city2', 'airport_1', 'airport_2', 'fare']
duplicates_key = df.duplicated(subset=key_columns).sum()
print(f"Duplicates based on key columns {key_columns}: {duplicates_key}")
print(f"Percentage: {(duplicates_key / len(df)) * 100:.2f}%")

if duplicates_key > 0:
    print("\nSample duplicate rows:")
    duplicate_rows = df[df.duplicated(subset=key_columns, keep=False)]
    print(duplicate_rows.head(10))
```

```python
# Data Quality Assessment - Outliers in Fare
import numpy as np

print("=== OUTLIER ANALYSIS - FARE ===")
fare_stats = df['fare'].describe()
print("Fare statistics:")
print(fare_stats)

# IQR method for outlier detection
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['fare'] < lower_bound) | (df['fare'] > upper_bound)]
print(f"\nOutliers (IQR method):")
print(f"Lower bound: {lower_bound:.2f}")
print(f"Upper bound: {upper_bound:.2f}")
print(f"Number of outliers: {len(outliers)}")
print(f"Percentage of outliers: {(len(outliers) / len(df)) * 100:.2f}%")

# Show some extreme values
print(f"\nExtreme fare values:")
print(f"Minimum fare: ${df['fare'].min():.2f}")
print(f"Maximum fare: ${df['fare'].max():.2f}")
print(f"Mean fare: ${df['fare'].mean():.2f}")
print(f"Median fare: ${df['fare'].median():.2f}")
```

```python
# Data Consistency Check
print("=== DATA CONSISTENCY CHECK ===")

# Check for negative fares
negative_fares = df[df['fare'] < 0]
print(f"Negative fares: {len(negative_fares)}")

# Check for zero fares
zero_fares = df[df['fare'] == 0]
print(f"Zero fares: {len(zero_fares)}")

# Check year range
print(f"\nYear range: {df['Year'].min()} - {df['Year'].max()}")
print(f"Unique years: {sorted(df['Year'].unique())}")

# Check quarter values
print(f"\nQuarter values: {sorted(df['quarter'].unique())}")

# Check for inconsistent quarter values
invalid_quarters = df[(df['quarter'] < 1) | (df['quarter'] > 4)]
print(f"Invalid quarter values: {len(invalid_quarters)}")

# Check for missing airport codes
missing_airport1 = df['airport_1'].isnull().sum()
missing_airport2 = df['airport_2'].isnull().sum()
print(f"\nMissing airport_1 codes: {missing_airport1}")
print(f"Missing airport_2 codes: {missing_airport2}")

# Check for missing city names
missing_city1 = df['city1'].isnull().sum()
missing_city2 = df['city2'].isnull().sum()
print(f"Missing city1 names: {missing_city1}")
print(f"Missing city2 names: {missing_city2}")
```

```python
# Temporal Analysis - Year and Quarter Trends
print("=== TEMPORAL ANALYSIS ===")

# Year distribution
print("Year distribution:")
year_counts = df['Year'].value_counts().sort_index()
print(year_counts)

# Quarter distribution
print("\nQuarter distribution:")
quarter_counts = df['quarter'].value_counts().sort_index()
print(quarter_counts)

# Year-Quarter combinations
print("\nYear-Quarter combinations:")
year_quarter = df.groupby(['Year', 'quarter']).size().reset_index(name='count')
print(year_quarter.head(20))

# Check for missing year-quarter combinations
all_years = range(df['Year'].min(), df['Year'].max() + 1)
all_quarters = [1, 2, 3, 4]
expected_combinations = [(year, quarter) for year in all_years for quarter in all_quarters]
actual_combinations = set(zip(df['Year'], df['quarter']))
missing_combinations = set(expected_combinations) - actual_combinations

print(f"\nMissing year-quarter combinations: {len(missing_combinations)}")
if missing_combinations:
    print("Missing combinations:")
    for combo in sorted(missing_combinations):
        print(f"  {combo[0]} Q{combo[1]}")
```

```python
# Fare Trends Over Time
print("=== FARE TRENDS OVER TIME ===")

# Average fare by year
yearly_fare = df.groupby('Year')['fare'].agg(['mean', 'median', 'std', 'count']).round(2)
print("Yearly fare statistics:")
print(yearly_fare)

# Average fare by quarter
quarterly_fare = df.groupby('quarter')['fare'].agg(['mean', 'median', 'std', 'count']).round(2)
print("\nQuarterly fare statistics:")
print(quarterly_fare)

# Year-Quarter fare analysis
year_quarter_fare = df.groupby(['Year', 'quarter'])['fare'].agg(['mean', 'median', 'count']).round(2)
print("\nYear-Quarter fare analysis (first 20 rows):")
print(year_quarter_fare.head(20))

# Fare growth rate
if len(yearly_fare) > 1:
    first_year_avg = yearly_fare.iloc[0]['mean']
    last_year_avg = yearly_fare.iloc[-1]['mean']
    growth_rate = ((last_year_avg - first_year_avg) / first_year_avg) * 100
    print(f"\nFare growth from {yearly_fare.index[0]} to {yearly_fare.index[-1]}: {growth_rate:.2f}%")
```

```python
# Geographic Analysis - Cities and Airports
print("=== GEOGRAPHIC ANALYSIS ===")

# Top cities by flight count
print("Top 20 cities by flight count:")
city1_counts = df['city1'].value_counts()
city2_counts = df['city2'].value_counts()

# Combine both origin and destination cities
all_cities = pd.concat([df['city1'], df['city2']])
city_totals = all_cities.value_counts()
print(city_totals.head(20))

# Top airports by flight count
print("\nTop 20 airports by flight count:")
airport1_counts = df['airport_1'].value_counts()
airport2_counts = df['airport_2'].value_counts()

# Combine both origin and destination airports
all_airports = pd.concat([df['airport_1'], df['airport_2']])
airport_totals = all_airports.value_counts()
print(airport_totals.head(20))

# Unique counts
print(f"\nUnique cities: {all_cities.nunique()}")
print(f"Unique airports: {all_airports.nunique()}")
print(f"Unique city pairs: {df.groupby(['city1', 'city2']).size().shape[0]}")
print(f"Unique airport pairs: {df.groupby(['airport_1', 'airport_2']).size().shape[0]}")
```

```python
# Route Analysis - Popular Routes
print("=== ROUTE ANALYSIS ===")

# Most popular routes (city pairs)
print("Top 20 most popular routes (city pairs):")
route_counts = df.groupby(['city1', 'city2']).size().reset_index(name='flight_count')
route_counts = route_counts.sort_values('flight_count', ascending=False)
print(route_counts.head(20))

# Most popular routes (airport pairs)
print("\nTop 20 most popular routes (airport pairs):")
airport_route_counts = df.groupby(['airport_1', 'airport_2']).size().reset_index(name='flight_count')
airport_route_counts = airport_route_counts.sort_values('flight_count', ascending=False)
print(airport_route_counts.head(20))

# Bidirectional routes (check if A->B and B->A both exist)
print("\nBidirectional route analysis:")
df['route_key'] = df.apply(lambda x: tuple(sorted([x['city1'], x['city2']])), axis=1)
bidirectional_routes = df.groupby('route_key').size()
bidirectional_count = (bidirectional_routes > 1).sum()
print(f"Routes with bidirectional traffic: {bidirectional_count}")

# Average fare by route
print("\nTop 20 most expensive routes (by average fare):")
route_fares = df.groupby(['city1', 'city2'])['fare'].agg(['mean', 'count']).round(2)
route_fares = route_fares[route_fares['count'] >= 5]  # Filter routes with at least 5 flights
route_fares = route_fares.sort_values('mean', ascending=False)
print(route_fares.head(20))
```

```python
# Fare Analysis - Pricing Patterns
print("=== FARE ANALYSIS ===")

# Fare distribution
print("Fare distribution analysis:")
fare_percentiles = df['fare'].quantile([0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99])
print("Fare percentiles:")
for p, value in fare_percentiles.items():
    print(f"{p*100:4.0f}th percentile: ${value:.2f}")

# Fare by distance (if available)
if 'distance' in df.columns:
    print("\nFare vs Distance analysis:")
    fare_distance = df[['fare', 'distance']].corr()
    print(f"Correlation between fare and distance: {fare_distance.iloc[0,1]:.3f}")
    
    # Average fare by distance ranges
    df['distance_range'] = pd.cut(df['distance'], bins=5, labels=['Very Short', 'Short', 'Medium', 'Long', 'Very Long'])
    distance_fare = df.groupby('distance_range')['fare'].agg(['mean', 'median', 'count']).round(2)
    print("\nFare by distance range:")
    print(distance_fare)

# Fare by carrier type
print("\nFare analysis by carrier type:")
if 'carrier_lg' in df.columns and 'carrier_low' in df.columns:
    # Large carrier fares
    large_carrier_fares = df[df['carrier_lg'].notna()]['fare_lg']
    low_carrier_fares = df[df['carrier_low'].notna()]['fare_low']
    
    print(f"Large carrier average fare: ${large_carrier_fares.mean():.2f}")
    print(f"Low-cost carrier average fare: ${low_carrier_fares.mean():.2f}")
    print(f"Price difference: ${large_carrier_fares.mean() - low_carrier_fares.mean():.2f}")
    print(f"Price difference percentage: {((large_carrier_fares.mean() - low_carrier_fares.mean()) / low_carrier_fares.mean()) * 100:.1f}%")
```

```python
# Fare Seasonality Analysis
print("=== FARE SEASONALITY ANALYSIS ===")

# Quarterly fare patterns
quarterly_fare_stats = df.groupby('quarter')['fare'].agg(['mean', 'median', 'std', 'count']).round(2)
print("Quarterly fare statistics:")
print(quarterly_fare_stats)

# Identify seasonal patterns
if len(quarterly_fare_stats) > 0:
    q1_avg = quarterly_fare_stats.loc[1, 'mean']
    q2_avg = quarterly_fare_stats.loc[2, 'mean'] 
    q3_avg = quarterly_fare_stats.loc[3, 'mean']
    q4_avg = quarterly_fare_stats.loc[4, 'mean']
    
    print(f"\nSeasonal fare comparison:")
    print(f"Q1 (Winter) average: ${q1_avg:.2f}")
    print(f"Q2 (Spring) average: ${q2_avg:.2f}")
    print(f"Q3 (Summer) average: ${q3_avg:.2f}")
    print(f"Q4 (Fall) average: ${q4_avg:.2f}")
    
    # Find highest and lowest fare quarters
    max_quarter = quarterly_fare_stats['mean'].idxmax()
    min_quarter = quarterly_fare_stats['mean'].idxmin()
    print(f"\nHighest fare quarter: Q{max_quarter} (${quarterly_fare_stats.loc[max_quarter, 'mean']:.2f})")
    print(f"Lowest fare quarter: Q{min_quarter} (${quarterly_fare_stats.loc[min_quarter, 'mean']:.2f})")
    
    seasonal_variation = ((quarterly_fare_stats['mean'].max() - quarterly_fare_stats['mean'].min()) / quarterly_fare_stats['mean'].min()) * 100
    print(f"Seasonal variation: {seasonal_variation:.1f}%")
```

```python
# Carrier Analysis - Market Share and Performance
print("=== CARRIER ANALYSIS ===")

# Large carrier analysis
if 'carrier_lg' in df.columns:
    print("Large carrier analysis:")
    large_carrier_stats = df.groupby('carrier_lg').agg({
        'fare_lg': ['mean', 'median', 'count'],
        'large_ms': 'mean'
    }).round(2)
    large_carrier_stats.columns = ['avg_fare', 'median_fare', 'flight_count', 'avg_market_share']
    large_carrier_stats = large_carrier_stats.sort_values('flight_count', ascending=False)
    print(large_carrier_stats.head(10))

# Low-cost carrier analysis
if 'carrier_low' in df.columns:
    print("\nLow-cost carrier analysis:")
    low_carrier_stats = df.groupby('carrier_low').agg({
        'fare_low': ['mean', 'median', 'count'],
        'lf_ms': 'mean'
    }).round(2)
    low_carrier_stats.columns = ['avg_fare', 'median_fare', 'flight_count', 'avg_load_factor']
    low_carrier_stats = low_carrier_stats.sort_values('flight_count', ascending=False)
    print(low_carrier_stats.head(10))

# Market share analysis
print("\nMarket share analysis:")
if 'large_ms' in df.columns:
    avg_market_share = df['large_ms'].mean()
    print(f"Average large carrier market share: {avg_market_share:.3f}")

if 'lf_ms' in df.columns:
    avg_load_factor = df['lf_ms'].mean()
    print(f"Average load factor: {avg_load_factor:.3f}")

# Carrier competition analysis
print("\nCarrier competition analysis:")
if 'carrier_lg' in df.columns and 'carrier_low' in df.columns:
    # Routes with both large and low-cost carriers
    competition_routes = df[(df['carrier_lg'].notna()) & (df['carrier_low'].notna())]
    print(f"Routes with both large and low-cost carriers: {len(competition_routes)}")
    
    if len(competition_routes) > 0:
        price_competition = competition_routes['fare_lg'] - competition_routes['fare_low']
        print(f"Average price difference (large - low-cost): ${price_competition.mean():.2f}")
        print(f"Median price difference: ${price_competition.median():.2f}")
```

```python
# Carrier Trends Over Time
print("=== CARRIER TRENDS OVER TIME ===")

# Large carrier market share over time
if 'carrier_lg' in df.columns and 'large_ms' in df.columns:
    print("Large carrier market share trends:")
    yearly_market_share = df.groupby('Year')['large_ms'].agg(['mean', 'median', 'std']).round(3)
    print(yearly_market_share)
    
    # Market share growth
    if len(yearly_market_share) > 1:
        first_year_share = yearly_market_share.iloc[0]['mean']
        last_year_share = yearly_market_share.iloc[-1]['mean']
        share_change = last_year_share - first_year_share
        print(f"\nMarket share change from {yearly_market_share.index[0]} to {yearly_market_share.index[-1]}: {share_change:.3f}")

# Load factor trends over time
if 'lf_ms' in df.columns:
    print("\nLoad factor trends:")
    yearly_load_factor = df.groupby('Year')['lf_ms'].agg(['mean', 'median', 'std']).round(3)
    print(yearly_load_factor)
    
    # Load factor growth
    if len(yearly_load_factor) > 1:
        first_year_lf = yearly_load_factor.iloc[0]['mean']
        last_year_lf = yearly_load_factor.iloc[-1]['mean']
        lf_change = last_year_lf - first_year_lf
        print(f"\nLoad factor change from {yearly_load_factor.index[0]} to {yearly_load_factor.index[-1]}: {lf_change:.3f}")

# Carrier performance by quarter
print("\nCarrier performance by quarter:")
if 'large_ms' in df.columns:
    quarterly_market_share = df.groupby('quarter')['large_ms'].mean().round(3)
    print("Quarterly market share:")
    print(quarterly_market_share)

if 'lf_ms' in df.columns:
    quarterly_load_factor = df.groupby('quarter')['lf_ms'].mean().round(3)
    print("\nQuarterly load factor:")
    print(quarterly_load_factor)
```

```python
# Route Characteristics Analysis
print("=== ROUTE CHARACTERISTICS ANALYSIS ===")

# Route frequency distribution
print("Route frequency distribution:")
route_frequencies = df.groupby(['city1', 'city2']).size()
frequency_distribution = route_frequencies.value_counts().sort_index()
print("Number of routes by frequency:")
print(frequency_distribution.head(20))

# Route diversity analysis
print(f"\nRoute diversity metrics:")
print(f"Total unique routes: {len(route_frequencies)}")
print(f"Routes with 1 flight: {(route_frequencies == 1).sum()}")
print(f"Routes with 2-5 flights: {((route_frequencies >= 2) & (route_frequencies <= 5)).sum()}")
print(f"Routes with 6-20 flights: {((route_frequencies >= 6) & (route_frequencies <= 20)).sum()}")
print(f"Routes with 21+ flights: {(route_frequencies >= 21).sum()}")

# Most and least frequent routes
print(f"\nMost frequent routes:")
most_frequent = route_frequencies.nlargest(10)
for route, count in most_frequent.items():
    print(f"{route[0]} -> {route[1]}: {count} flights")

print(f"\nLeast frequent routes (sample):")
least_frequent = route_frequencies.nsmallest(10)
for route, count in least_frequent.items():
    print(f"{route[0]} -> {route[1]}: {count} flights")

# Route stability over time
print(f"\nRoute stability analysis:")
# Routes that appear in multiple years
route_years = df.groupby(['city1', 'city2'])['Year'].nunique()
stable_routes = route_years[route_years > 1]
print(f"Routes appearing in multiple years: {len(stable_routes)}")
print(f"Most stable routes (appearing in most years):")
most_stable = stable_routes.nlargest(10)
for route, years in most_stable.items():
    print(f"{route[0]} -> {route[1]}: {years} years")
```

```python
# Route Performance Analysis
print("=== ROUTE PERFORMANCE ANALYSIS ===")

# Route profitability analysis (using fare as proxy)
route_performance = df.groupby(['city1', 'city2']).agg({
    'fare': ['mean', 'median', 'std', 'count'],
    'Year': 'nunique'
}).round(2)

route_performance.columns = ['avg_fare', 'median_fare', 'fare_std', 'flight_count', 'years_active']
route_performance = route_performance.sort_values('avg_fare', ascending=False)

print("Top 20 most expensive routes (by average fare):")
print(route_performance.head(20))

print("\nTop 20 most affordable routes (by average fare):")
print(route_performance.tail(20))

# Route volume vs fare analysis
print("\nRoute volume vs fare analysis:")
# High volume routes (top 20% by flight count)
volume_threshold = route_performance['flight_count'].quantile(0.8)
high_volume_routes = route_performance[route_performance['flight_count'] >= volume_threshold]
print(f"High volume routes (≥{volume_threshold:.0f} flights): {len(high_volume_routes)}")
print(f"Average fare for high volume routes: ${high_volume_routes['avg_fare'].mean():.2f}")

# Low volume routes (bottom 20% by flight count)
low_volume_threshold = route_performance['flight_count'].quantile(0.2)
low_volume_routes = route_performance[route_performance['flight_count'] <= low_volume_threshold]
print(f"Low volume routes (≤{low_volume_threshold:.0f} flights): {len(low_volume_routes)}")
print(f"Average fare for low volume routes: ${low_volume_routes['avg_fare'].mean():.2f}")

# Route consistency analysis
print("\nRoute fare consistency analysis:")
route_performance['fare_cv'] = route_performance['fare_std'] / route_performance['avg_fare']  # Coefficient of variation
consistent_routes = route_performance[route_performance['fare_cv'] < 0.3]  # Low variation
print(f"Consistent fare routes (CV < 0.3): {len(consistent_routes)}")
print(f"Average fare for consistent routes: ${consistent_routes['avg_fare'].mean():.2f}")

volatile_routes = route_performance[route_performance['fare_cv'] > 0.5]  # High variation
print(f"Volatile fare routes (CV > 0.5): {len(volatile_routes)}")
print(f"Average fare for volatile routes: ${volatile_routes['avg_fare'].mean():.2f}")
```
